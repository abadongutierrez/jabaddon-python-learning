```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "laptop_pricing_dataset_mod1.csv"
```

```python
df = pd.read_csv(file_path, header=0)
```

```python
df.head(10)
```

```python
df[['Screen_Size_cm']] = np.round(df[['Screen_Size_cm']],2)
df.head()
```

```python
# Replace "?" with NaN values

import numpy as np

df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
```

```python
missing_data.value_counts()
```

```python
missing_data["Weight_kg"].value_counts()
```

```python
# get mean value for "Weight_kg" column
mean_weight = df["Weight_kg"].astype("float").mean(axis=0)
print("Mean weight:", mean_weight)
# replace NaN values in "Weight_kg" column with mean value
df["Weight_kg"].replace(np.nan, mean_weight, inplace=True)
df.head(5)
```

```python
df.isnull()["Weight_kg"].value_counts()
```

```python
df["Screen_Size_cm"].isnull().value_counts()
```

```python
# get most frequent value for "Screen_Size_cm" column
most_frequent_size = df["Screen_Size_cm"].astype("float").mode()[0]
print("Most frequent screen size:", most_frequent_size)
```

```python
# Replace "NaN" values with the most frequent value in "Screen_Size_cm" column
df["Screen_Size_cm"].replace(np.nan, most_frequent_size, inplace=True)
df.head(5)
```

```python
df.dtypes
```

```python
# ensure Screen_Size_cm and Weight_kg are float
df[["Screen_Size_cm", "Weight_kg"]] = df[["Screen_Size_cm", "Weight_kg"]].astype("float")
```

```python
# Data standardization
# 1. Lets create Screen_Size_inch column from Screen_Size_cm
# 1 inch = 2.54 cm
df["Screen_Size_inch"] = np.round(df["Screen_Size_cm"] / 2.54, 2)

# 2. Lets create Weight_pounds column from Weight_kg
# 1 kg = 2.20462 pounds
df["Weight_pounds"] = np.round(df["Weight_kg"] * 2.20462, 2)
```

```python
df[["Screen_Size_cm", "Screen_Size_inch", "Weight_kg", "Weight_pounds"]].head(10)
```

```python
df.columns
```

```python
# Data normalization

# Lets normalize CPU_Frequency column
df["CPU_frequency"] = df["CPU_frequency"].astype("float")
df["CPU_frequency"] = df["CPU_frequency"] / df["CPU_frequency"].max()
```

```python
df["CPU_frequency"].head()
```

```python
# Binning

# Lets create bings for "Price"
# Bings should be "Low", "Medium", "High"
bins = np.linspace(min(df["Price"]), max(df["Price"]), 4)
group_names = ["Low", "Medium", "High"]
df['Price-binned'] = pd.cut(df['Price'], bins, labels=group_names, include_lowest=True )
df[["Price", "Price_binned"]].head(10)
```

```python
#Indicator Variable: Screen
dummy_variable_1 = pd.get_dummies(df["Screen"])
dummy_variable_1.rename(columns={'IPS Panel':'Screen-IPS_panel', 'Full HD':'Screen-Full_HD'}, inplace=True)
df = pd.concat([df, dummy_variable_1], axis=1)

# drop original column "Screen" from "df"
df.drop("Screen", axis = 1, inplace=True)
```
